```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
fig , axis = plt.subplots(2,2 , figsize = (16,9))
sns.histplot(df["sepal_length"] , ax = axis[0,0])
sns.histplot(df["sepal_width"] , ax = axis[0,1])
sns.histplot(df["petal_length"] , ax = axis[1,0])
sns.histplot(df["petal_width"] , ax = axis[1,1])
```

```python
#part 2
import matplotlib.pyplot as plt
fig , axis = plt.subplots(2,2 , figsize = (16,9))
sns.boxplot(y = "sepal_length" , x="species",data = df , ax = axis[0,0])
sns.boxplot(y = "sepal_width" , x="species",data = df , ax = axis[0,1])
sns.boxplot(y = "petal_length" , x="species",data = df , ax = axis[1,0])
sns.boxplot(y = "petal_length" , x="species",data = df , ax = axis[1,1])
```
